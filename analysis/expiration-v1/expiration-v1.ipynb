```python
from mypy.datautil import flatten
import ConfigParser
import numpy as np
import matplotlib.pyplot as plt
import json
import random as rnd
import pandas as pd
from scipy.stats import ttest_ind, beta
from exputil import load_from_mysql
from gambles import expected_value
from expiration import *
from plotting import *

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.stats.mstats import kruskalwallis

j = .15
def jit(values):
    return [v+rnd.uniform(-j,j) for v in values]

MAXSAMP = 26
N_GAMES = 12

DATA_DIR = "/Users/markant/data/CompetitiveSampling/expiration-v1"
ANALYSIS_DIR = "/Users/markant/code/CompetitiveSampling/analysis/expiration-v1"

Config = ConfigParser.ConfigParser()
Config.read("/Users/markant/.secrets")
DBURL = Config.get("Experiment Data", "smashdburl")

"""
def rawdata(s):
    with open('%s/%s.json' % (DATA_DIR, s), 'r') as f:
        j = json.loads(f.read())
    return j
"""

def rawdata(s):
    return json.loads(DATA[s])

def trialdata(s):
    return [t['trialdata'] for t in rawdata(s)['data']]

def trialdata_with_times(s):
    td = []
    for t in [[d['trialdata'], d['dateTime']] for d in rawdata(s)['data']]:
        tdi = []
        for entry in t[0]:
            tdi.append(entry)
        tdi.append(t[1])
        td.append(tdi)
    return td

def questiondata(sid): 
    return rawdata(sid)['questiondata']

def eventdata(sid): 
    return rawdata(sid)['eventdata']


def parsegames(alldata):
    data = filter(lambda t: t[0]=='game', alldata)
    
    n_practice_games = len(filter(lambda t: t[2]=='practice' and t[3]==True, data))
    
    games = {i:[] for i in range(N_GAMES)}
    practice = True
    practice_counter = 0
    for i in range(len(data)):
        gameind = data[i][1]
        
        if not practice:
            #print data[i]
            games[gameind].append(data[i])
            
        elif data[i][2]=='practice' and data[i][3]==True:
            practice_counter += 1
        
        elif practice and data[i][3]=='choice' and practice_counter==n_practice_games:
            practice = False
            
    return [Game(games[i]) for i in range(N_GAMES)]

    
class Game:
    
    def __init__(self, data):
        
        self.data = data
        self.gindex = data[0][1]
        self.gamble_id = filter(lambda t: t[2]=='gamble' and t[3]=='id', data)[0][4]
        
        self.options = {}
        self.options['A'] = filter(lambda t: t[2]=='option' and t[3]=='A', data)[0][4:7]
        self.options['B'] = filter(lambda t: t[2]=='option' and t[3]=='B', data)[0][4:7]
        self.samples = [s[4:] for s in filter(lambda t: t[3]=='sample', data)]
        self.n_samples = len(self.samples)
        self.expiration = int(filter(lambda t: t[2]=='expiration', data)[0][3])

        self.choice = filter(lambda t: t[3]=='choice', data)[0][4]
        
        gd = self.data
        sampleind = filter(lambda i: gd[i][3]=='sample', range(len(gd)))
        self.sample_rt = [gd[i][-1] - gd[i-1][-1] for i in sampleind]

        choiceind = filter(lambda i: gd[i][3]=='choice', range(len(gd)))[0]
        self.choice_rt = gd[choiceind][-1] - gd[choiceind-1][-1]

        if expected_value(self.options['A']) > expected_value(self.options['B']):
            self.H = 'A'
            self.L = 'B'
        else:
            self.H = 'B'
            self.L = 'A'
        if self.choice == self.H:
            self.chose_H = True
        else:
            self.chose_H = False
        
        pd = filter(lambda t: t[2]=='planning_decision', data)
        if len(pd)>0: 
            self.planning_decision = pd[0][3]
        else:
            self.planning_decision = None
            
        ex = filter(lambda t: t[3]=='expired', data)
        if len(ex)>0:
            self.observed_expiration = ex[0][2]
        else:
            self.observed_expiration = None


class Subject:
    
    def __init__(self, sid):
        
        self.sid = sid
        
        try:
            self.data = trialdata_with_times(self.sid)
        except:
            self.data = None
            print 'failed to load data at %s/%s.json' % (DATA_DIR, self.sid)
        
        if self.data != None:
            self.condition = int(filter(lambda t: t[0]=='condition', self.data)[0][1])
            self.condition_tbt = int(filter(lambda t: t[0]=='condition_tbt', self.data)[0][1])
            self.condition_exp = int(filter(lambda t: t[0]=='condition_exp', self.data)[0][1])
            
            self.bonus = filter(lambda t: len(t)>2 and t[2]=='final_bonus', self.data)[0][3] 
            self.games = parsegames(self.data)
            
            self.questiondata = questiondata(self.sid)
            
            self.focuschanges = len(filter(lambda ev: ev['eventtype']=='focus' and ev['value']=='off', eventdata(self.sid)))
```

```python
def sample_feedback(samples):
    obs = []
    for s in samples:
        if s[0]=='A':
            if s[1] > 0:
                obs.append('Apos')
            else:
                obs.append('Aneg')
        else:
            if s[1] > 0:
                obs.append('Bpos')
            else:
                obs.append('Bneg')
    return np.array(obs)

def model_game(sid, gameind):
    g = Subject(sid).games[gameind]
    obs = sample_feedback(g.samples)    
    models = []
    
    #! only doing last trial right now!
    for n in range(len(obs)-1,len(obs)):
        counts = {'Apos': np.sum(obs[:n]=='Apos'),
                  'Aneg': np.sum(obs[:n]=='Aneg'),
                  'Bpos': np.sum(obs[:n]=='Bpos'),
                  'Bneg': np.sum(obs[:n]=='Bneg')}
        
        if counts['Apos'] > 0:
            Apos_min, Apos_max = g.options['A'][0], g.options['A'][0]
        else:
            Apos_min, Apos_max = 1, 100
        
        if counts['Aneg'] > 0:
            Aneg_min, Aneg_max = g.options['A'][1], g.options['A'][1]
        else:
            Aneg_min, Aneg_max = -100, -1
        
        if counts['Bpos'] > 0:
            Bpos_min, Bpos_max = g.options['B'][0], g.options['B'][0]
        else:
            Bpos_min, Bpos_max = 1, 100
        
        if counts['Bneg'] > 0:
            Bneg_min, Bneg_max = g.options['B'][1], g.options['B'][1]
        else:
            Bneg_min, Bneg_max = -100, -1
        
        Apos = pm.DiscreteUniform('Apos', lower=Apos_min, upper=Apos_max)
        Aneg = pm.DiscreteUniform('Aneg', lower=Aneg_min, upper=Aneg_max)
        Apr  = pm.Beta('Apr', 1 + counts['Apos'], 1 + counts['Aneg'])
        Bpos = pm.DiscreteUniform('Bpos', lower=Bpos_min, upper=Bpos_max)
        Bneg = pm.DiscreteUniform('Bneg', lower=Bneg_min, upper=Bneg_max)
        Bpr  = pm.Beta('Bpr', 1 + counts['Bpos'], 1 + counts['Bneg'])
        
        @pm.deterministic
        def EVA(pos=Apos, neg=Aneg, p=Apr):
            return (p * pos) + (1 - p) * neg
        
        @pm.deterministic
        def EVB(pos=Bpos, neg=Bneg, p=Bpr):
            return (p * pos) + (1 - p) * neg
        
        @pm.deterministic
        def pAisH(eva=EVA, evb=EVB):
            if eva > evb:
                return True
            else:
                return False
        
        nsamp = 5000.
        models.append(pm.MCMC([Apos, Aneg, Apr, Bpos, Bneg, Bpr, EVA, EVB, pAisH]))
        models[-1].sample(nsamp, burn=500)
        
    return models
```
