```python
from mypy.datautil import flatten
import ConfigParser
import numpy as np
import matplotlib.pyplot as plt
import json
import random as rnd
import pandas as pd
from scipy.stats import ttest_ind, beta
from exputil import load_from_mysql
from gambles import expected_value
from expiration import *
from plotting import *
from copy import deepcopy

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.stats.mstats import kruskalwallis
from scipy.stats import mannwhitneyu

# for bayesian model
import pymc as pm

j = .15
def jit(values):
    return [v+rnd.uniform(-j,j) for v in values]

MAXSAMP = 26
N_GAMES = 12

EXP_LABELS = ['No expiration', 'Uniform', 'Normal', 'Exponential', 'Exponential-2']

DATA_DIR = "/Users/markant/data/CompetitiveSampling/expiration-v1"
ANALYSIS_DIR = "/Users/markant/code/CompetitiveSampling/analysis/expiration-v1"

Config = ConfigParser.ConfigParser()
Config.read("/Users/markant/.secrets")
#Config.read("/Users/doug/.secrets")
DBURL = Config.get("Experiment Data", "smashdburl")

"""
def rawdata(s):
    with open('%s/%s.json' % (DATA_DIR, s), 'r') as f:
        j = json.loads(f.read())
    return j
"""

def rawdata(s):
    return json.loads(DATA[s])

def trialdata(s):
    return [t['trialdata'] for t in rawdata(s)['data']]

def trialdata_with_times(s):
    td = []
    for t in [[d['trialdata'], d['dateTime']] for d in rawdata(s)['data']]:
        tdi = []
        for entry in t[0]:
            tdi.append(entry)
        tdi.append(t[1])
        td.append(tdi)
    return td

def questiondata(sid): 
    return rawdata(sid)['questiondata']

def eventdata(sid): 
    return rawdata(sid)['eventdata']


def parsegames(alldata):
    data = filter(lambda t: t[0]=='game', alldata)
    
    n_practice_games = len(filter(lambda t: t[2]=='practice' and t[3]==True, data))
    
    games = {i:[] for i in range(N_GAMES)}
    practice = True
    practice_counter = 0
    for i in range(len(data)):
        gameind = data[i][1]
        
        if not practice:
            #print data[i]
            games[gameind].append(data[i])
            
        elif data[i][2]=='practice' and data[i][3]==True:
            practice_counter += 1
        
        elif practice and data[i][3]=='choice' and practice_counter==n_practice_games:
            practice = False
            
    return [Game(games[i]) for i in range(N_GAMES)]

    
class Game:
    
    def __init__(self, data):
        
        self.data = data
        self.gindex = data[0][1]
        self.gamble_id = filter(lambda t: t[2]=='gamble' and t[3]=='id', data)[0][4]
        
        self.options = {}
        self.options['A'] = filter(lambda t: t[2]=='option' and t[3]=='A', data)[0][4:7]
        self.options['B'] = filter(lambda t: t[2]=='option' and t[3]=='B', data)[0][4:7]
        self.samples = [s[4:] for s in filter(lambda t: t[3]=='sample', data)]
        self.n_samples = len(self.samples)
        self.expiration = int(filter(lambda t: t[2]=='expiration', data)[0][3])

        self.choice = filter(lambda t: t[3]=='choice', data)[0][4]
        
        gd = self.data
        sampleind = filter(lambda i: gd[i][3]=='sample', range(len(gd)))
        self.sample_rt = [gd[i][-1] - gd[i-1][-1] for i in sampleind]

        choiceind = filter(lambda i: gd[i][3]=='choice', range(len(gd)))[0]
        self.choice_rt = gd[choiceind][-1] - gd[choiceind-1][-1]

        if expected_value(self.options['A']) > expected_value(self.options['B']):
            self.H = 'A'
            self.L = 'B'
        else:
            self.H = 'B'
            self.L = 'A'
        if self.choice == self.H:
            self.chose_H = True
        else:
            self.chose_H = False
        
        pd = filter(lambda t: t[2]=='planning_decision', data)
        if len(pd)>0: 
            self.planning_decision = pd[0][3]
        else:
            self.planning_decision = None
            
        ex = filter(lambda t: t[3]=='expired', data)
        if len(ex)>0:
            self.observed_expiration = ex[0][2]
        else:
            self.observed_expiration = None


class Subject:
    
    def __init__(self, sid):
        
        self.sid = sid
        
        try:
            self.data = trialdata_with_times(self.sid)
        except:
            self.data = None
            print 'failed to load data at %s/%s.json' % (DATA_DIR, self.sid)
        
        if self.data != None:
            self.condition = int(filter(lambda t: t[0]=='condition', self.data)[0][1])
            self.condition_tbt = int(filter(lambda t: t[0]=='condition_tbt', self.data)[0][1])
            self.condition_exp = int(filter(lambda t: t[0]=='condition_exp', self.data)[0][1])
            
            self.bonus = filter(lambda t: len(t)>2 and t[2]=='final_bonus', self.data)[0][3] 
            self.games = parsegames(self.data)
            
            self.questiondata = questiondata(self.sid)
            
            self.focuschanges = len(filter(lambda ev: ev['eventtype']=='focus' and ev['value']=='off', eventdata(self.sid)))
```

```python
s = Subject(0)
g = s.games[0]
gambles = [{'H': tuple(g.options[g.H]), 'L': tuple(g.options[g.L])} for g in s.games]

pH =      np.zeros((len(gambles), maxsamp), float)
pH_unif = np.zeros((len(gambles), maxsamp), float)
pH_norm = np.zeros((len(gambles), maxsamp), float)
pH_exp  = np.zeros((len(gambles), maxsamp), float)
pH_exp2 = np.zeros((len(gambles), maxsamp), float)

for i in range(len(gambles)):
    opt = gambles[i]
    p_i = [prob_choose_H_all_allocations(opt, n) for n in range(1, MAXSAMP)]
    pH[i] = np.array(p_i)
    p_i, eg_i = expected_gain_given_discrete_expiration(opt, maxsamp, UNIFFREQ)
    pH_unif[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_expiration(opt, maxsamp, NORMFREQ)
    pH_norm[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_expiration(opt, maxsamp, EXPFREQ)
    pH_exp[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_expiration(opt, maxsamp, EXPFREQ2)
    pH_exp2[i] = p_i
```

# Adaptive model

```python
pexpire_unif = [UNIFFREQ[n]/np.sum(UNIFFREQ[n:]) for n in range(len(UNIFFREQ))]
pexpire_norm = [NORMFREQ[n]/np.sum(NORMFREQ[n:]) for n in range(len(NORMFREQ))]
pexpire_exp  = [EXPFREQ[n]/np.sum(EXPFREQ[n:]) for n in range(len(EXPFREQ))]

plt.plot(pexpire_unif)
plt.plot(pexpire_norm)
plt.plot(pexpire_exp)
plt.show()
```

```python
def sample_feedback(samples):
    obs = []
    for s in samples:
        if s[0]=='A':
            if s[1] > 0:
                obs.append('Apos')
            else:
                obs.append('Aneg')
        else:
            if s[1] > 0:
                obs.append('Bpos')
            else:
                obs.append('Bneg')
    return np.array(obs)

def outcome_bounds(observations):
    
    labels = obs_labels(observations)
    
    def get_value(l):
        return int(observations[where(labels==l)[0][0]][1])
            
    counts = {'Apos': np.sum(labels=='Apos'),
              'Aneg': np.sum(labels=='Aneg'),
              'Bpos': np.sum(labels=='Bpos'),
              'Bneg': np.sum(labels=='Bneg')}
        
    if counts['Apos'] > 0:
        value = get_value('Apos')
        Apos_min, Apos_max = value, value
    else:
        Apos_min, Apos_max = 1, 100
    
    if counts['Aneg'] > 0:
        value = get_value('Aneg')
        Aneg_min, Aneg_max = value, value
    else:
        Aneg_min, Aneg_max = -100, -1
    
    if counts['Bpos'] > 0:
        value = get_value('Bpos')
        Bpos_min, Bpos_max = value, value
    else:
        Bpos_min, Bpos_max = 1, 100
    
    if counts['Bneg'] > 0:
        value = get_value('Bneg')
        Bneg_min, Bneg_max = value, value
    else:
        Bneg_min, Bneg_max = -100, -1
    
    return [Apos_min, Apos_max, Aneg_min, Aneg_max, Bpos_min, Bpos_max, Bneg_min, Bneg_max]

def obs_labels(observations):
    return sample_feedback([s[:2] for s in observations])

def observed_hits(observations):
    hits_A = [1 if oi=='Apos' else 0 for oi in [o for o in observations if (o=='Aneg' or o=='Apos')]]
    hits_B = [1 if oi=='Bpos' else 0 for oi in [o for o in observations if (o=='Bneg' or o=='Bpos')]]
    return hits_A, hits_B
```

```python
def model_trial(observations):

    labels = obs_labels(observations)
    hits_A, hits_B = observed_hits(labels)
    
    pr_A   = pm.Beta('pr_A', 1, 1)
    obs_A  = pm.Bernoulli('obs_A', pr_A, value=hits_A, observed=True)
    next_A = pm.Bernoulli('next_A', pr_A)
    
    pr_B   = pm.Beta('pr_B', 1, 1)
    obs_B  = pm.Bernoulli('obs_B', pr_B, value=hits_B, observed=True)
    next_B = pm.Bernoulli('next_B', pr_B)
    
    [Apos_min, Apos_max, Aneg_min, Aneg_max, Bpos_min, Bpos_max, Bneg_min, Bneg_max] = outcome_bounds(observations)
    Apos = pm.DiscreteUniform('Apos', lower=Apos_min, upper=Apos_max)
    Aneg = pm.DiscreteUniform('Aneg', lower=Aneg_min, upper=Aneg_max)
    Bpos = pm.DiscreteUniform('Bpos', lower=Bpos_min, upper=Bpos_max)
    Bneg = pm.DiscreteUniform('Bneg', lower=Bneg_min, upper=Bneg_max)
    
    # predicting expected value
    @pm.deterministic
    def EVA(pos=Apos, neg=Aneg, p=pr_A):
        return (p * pos) + (1 - p) * neg
    
    @pm.deterministic
    def EVB(pos=Bpos, neg=Bneg, p=pr_B):
        return (p * pos) + (1 - p) * neg
    
    @pm.deterministic
    def EVD(eva=EVA, evb=EVB):
        return eva - evb

    @pm.deterministic
    def pAisH(eva=EVA, evb=EVB):
        if eva > evb:
            return True
        else:
            return False
    
    nsamp = 5000.
    m = pm.MCMC([obs_A, pr_A, next_A, Apos, Aneg, EVA, obs_B, pr_B, next_B, Bpos, Bneg, EVB, EVD, pAisH])
    m.sample(nsamp, burn=500)
    return m
```

### Probability gain model

```python
def estimate_probability_gain(sid, gameid, trial):
    
    # observations so far
    obs_t = [s[:2] for s in Subject(sid).games[gameid].samples][:trial]
    labels = obs_labels(obs_t)

    # model on current trial
    m = model_trial(obs_t)
    pAisH = np.sum(1*m.trace('pAisH')[:])/float(nsamp)
    pr_next_Apos = np.sum(1*m.trace('next_A')[:])/float(nsamp)
    pr_next_Bpos = np.sum(1*m.trace('next_B')[:])/float(nsamp)

    # step through possible outcomes of next observation
    pAisH_t2 = {}
    for l in ['Apos', 'Aneg', 'Bpos', 'Bneg']:
        ind = where(labels==l)[0]
        if len(ind)>0:
            # the value of this outcome is known
            obs_t2 = deepcopy(obs_t)
            obs_t2.append([unicode(l[0]), int(obs_t[ind[0]][1])])
            m_t2 = model_trial(obs_t2)
            pAisH_t2[l] = np.array([np.sum(m_t2.trace('pAisH')[:])/float(nsamp)])
        else:
            # the value of this outcome is not yet known
            pAisH_t2[l] = []
            bounds = [1, 101] if 'pos' in l else [-100, 0]
            for v in range(bounds[0], bounds[1]):
                obs_t2 = deepcopy(obs_t)
                obs_t2.append([unicode(l[0]), v])
                m_t2 = model_trial(obs_t2)
                pAisH_t2[l].append(np.sum(m_t2.trace('pAisH')[:])/float(nsamp))
            
    
    return pAisH, pr_next_Apos, pr_next_Bpos, pAisH_t2
```

```python
def model_game(sid, gameind):
    g = Subject(sid).games[gameind]
    obs = sample_feedback(g.samples)    
    models = []
    
    for n in range(len(obs)):
        counts = {'Apos': np.sum(obs[:n]=='Apos'),
                  'Aneg': np.sum(obs[:n]=='Aneg'),
                  'Bpos': np.sum(obs[:n]=='Bpos'),
                  'Bneg': np.sum(obs[:n]=='Bneg')}
        
        if counts['Apos'] > 0:
            Apos_min, Apos_max = g.options['A'][0], g.options['A'][0]
        else:
            Apos_min, Apos_max = 1, 100
        
        if counts['Aneg'] > 0:
            Aneg_min, Aneg_max = g.options['A'][1], g.options['A'][1]
        else:
            Aneg_min, Aneg_max = -100, -1
        
        if counts['Bpos'] > 0:
            Bpos_min, Bpos_max = g.options['B'][0], g.options['B'][0]
        else:
            Bpos_min, Bpos_max = 1, 100
        
        if counts['Bneg'] > 0:
            Bneg_min, Bneg_max = g.options['B'][1], g.options['B'][1]
        else:
            Bneg_min, Bneg_max = -100, -1
        
        Apos = pm.DiscreteUniform('Apos', lower=Apos_min, upper=Apos_max)
        Aneg = pm.DiscreteUniform('Aneg', lower=Aneg_min, upper=Aneg_max)
        Apr  = pm.Beta('Apr', 1 + counts['Apos'], 1 + counts['Aneg'])
        Bpos = pm.DiscreteUniform('Bpos', lower=Bpos_min, upper=Bpos_max)
        Bneg = pm.DiscreteUniform('Bneg', lower=Bneg_min, upper=Bneg_max)
        Bpr  = pm.Beta('Bpr', 1 + counts['Bpos'], 1 + counts['Bneg'])
        
        pApos= pm.Bernoulli('pApos', Apr)    
        pBpos= pm.Bernoulli('pBpos', Apr)            
        

        @pm.deterministic
        def EVA(pos=Apos, neg=Aneg, p=Apr):
            return (p * pos) + (1 - p) * neg
        
        @pm.deterministic
        def EVB(pos=Bpos, neg=Bneg, p=Bpr):
            return (p * pos) + (1 - p) * neg
        
        @pm.deterministic
        def pAisH(eva=EVA, evb=EVB):
            if eva > evb:
                return True
            else:
                return False
        
        
        nsamp = 5000.
        models.append(pm.MCMC([Apos, Aneg, Apr, Bpos, Bneg, Bpr, EVA, EVB, pAisH, pApos, pBpos]))
        models[-1].sample(nsamp, burn=500)
        
    return models
```
