```python
import numpy as np
import matplotlib.pyplot as plt
import json
import pandas as pd
from scipy.stats import ttest_ind
from exputil import load_from_mysql
from gambles import expected_value
from expiration import *
from plotting import *

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.stats.mstats import kruskalwallis

j = .15
def jit(values):
    return [v+random.uniform(-j,j) for v in values]


SUBJ = range(10)
MAXSAMP = 26
N_GAMES = 12

DATA_DIR = "/Users/markant/data/CompetitiveSampling/expiration-v1"
ANALYSIS_DIR = "/Users/markant/code/CompetitiveSampling/analysis/expiration-v1"

"""
def rawdata(s):
    with open('%s/%s.json' % (DATA_DIR, s), 'r') as f:
        j = json.loads(f.read())
    return j
"""

def rawdata(s):
    return json.loads(DATA[s])

def trialdata(s):
    return [t['trialdata'] for t in rawdata(s)['data']]

def questiondata(sid): 
    return rawdata(sid)['questiondata']

def parsegames(alldata):
    data = filter(lambda t: t[0]=='game', alldata)
    
    n_practice_games = len(filter(lambda t: t[2]=='practice' and t[3]==True, data))
    
    games = {i:[] for i in range(N_GAMES)}
    practice = True
    practice_counter = 0
    for i in range(len(data)):
        gameind = data[i][1]
        
        if not practice:
            #print data[i]
            games[gameind].append(data[i])
            
        elif data[i][2]=='practice' and data[i][3]==True:
            practice_counter += 1
        
        elif practice and data[i][3]=='choice' and practice_counter==n_practice_games:
            practice = False
            
    return [Game(games[i]) for i in range(N_GAMES)]

    
class Game:
    
    def __init__(self, data):
        
        self.data = data
        self.gindex = data[0][1]
        self.gamble_id = filter(lambda t: t[2]=='gamble' and t[3]=='id', data)[0][4]
        
        self.options = {}
        self.options['A'] = filter(lambda t: t[2]=='option' and t[3]=='A', data)[0][4:7]
        self.options['B'] = filter(lambda t: t[2]=='option' and t[3]=='B', data)[0][4:7]
        self.samples = [s[4:] for s in filter(lambda t: t[3]=='sample', data)]
        self.n_samples = len(self.samples)
        self.expiration = int(filter(lambda t: t[2]=='expiration', data)[0][3])

        self.choice = filter(lambda t: t[3]=='choice', data)[0][4]
        if expected_value(self.options['A']) > expected_value(self.options['B']):
            self.H = 'A'
            self.L = 'B'
        else:
            self.H = 'B'
            self.L = 'A'
        if self.choice == self.H:
            self.chose_H = True
        else:
            self.chose_H = False
        
        pd = filter(lambda t: t[2]=='planning_decision', data)
        if len(pd)>0: 
            self.planning_decision = pd[0][3]
        else:
            self.planning_decision = None
            
        ex = filter(lambda t: t[3]=='expired', data)
        if len(ex)>0:
            self.observed_expiration = ex[0][2]
        else:
            self.observed_expiration = None


class Subject:
    
    def __init__(self, sid):
        
        self.sid = sid
        
        try:
            self.data = trialdata(self.sid)
        except:
            self.data = None
            print 'failed to load data at %s/%s.json' % (DATA_DIR, self.sid)
        
        if self.data != None:
            self.condition = int(filter(lambda t: t[0]=='condition', self.data)[0][1])
            self.condition_tbt = int(filter(lambda t: t[0]=='condition_tbt', self.data)[0][1])
            self.condition_exp = int(filter(lambda t: t[0]=='condition_exp', self.data)[0][1])
            
            self.bonus = filter(lambda t: len(t)>2 and t[2]=='final_bonus', self.data)[0][3] 
            self.games = parsegames(self.data)
            
            self.questiondata = questiondata(self.sid)
```

Given each of these expiration conditions and the general structure of the gambles, what would the optimal *planned* number of samples be?

```python
maxsamp = 25
ngambles = 100
pH =      np.zeros((ngambles, maxsamp), float)
pH_unif = np.zeros((ngambles, maxsamp), float)
pH_norm = np.zeros((ngambles, maxsamp), float)
pH_exp  = np.zeros((ngambles, maxsamp), float)

for i in range(ngambles):
    if i%10==0: print i
    opt = generate_gamble_posneg()
    p_i = [prob_choose_H_all_allocations(opt, n) for n in range(1, MAXSAMP)]
    pH[i] = np.array(p_i)
    p_i, eg_i = expected_gain_given_discrete_uniform_expiration(opt, maxsamp)
    pH_unif[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_normal_expiration(opt, maxsamp)
    pH_norm[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_exponential_expiration(opt, maxsamp)
    pH_exp[i] = p_i
```

```python
x = range(1, maxsamp+1)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

for arr in [pH, pH_unif, pH_norm, pH_exp]:
    y = arr.mean(axis=0)
    err = arr.std(axis=0)/np.sqrt(arr.shape[0])
    ax.errorbar(x, y, yerr=err)

ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('total # samples')
ax.set_ylabel('p(choose H)')
ax.set_title('probability of getting H given uniform expiration')
plt.show()
```

Given each of these expiration conditions and the actual gambles that were used, what is the optimal *planned* number of samples?

```python
s = Subject(0)
g = s.games[0]
gambles = [{'H': tuple(g.options[g.H]), 'L': tuple(g.options[g.L])} for g in s.games]

pH =      np.zeros((len(gambles), maxsamp), float)
pH_unif = np.zeros((len(gambles), maxsamp), float)
pH_norm = np.zeros((len(gambles), maxsamp), float)
pH_exp  = np.zeros((len(gambles), maxsamp), float)

for i in range(len(gambles)):
    opt = gambles[i]
    p_i = [prob_choose_H_all_allocations(opt, n) for n in range(1, MAXSAMP)]
    pH[i] = np.array(p_i)
    p_i, eg_i = expected_gain_given_discrete_uniform_expiration(opt, maxsamp)
    pH_unif[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_normal_expiration(opt, maxsamp)
    pH_norm[i] = p_i
    p_i, eg_i = expected_gain_given_discrete_exponential_expiration(opt, maxsamp)
    pH_exp[i] = p_i
```

```python
x = range(1, maxsamp+1)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

for arr in [pH, pH_unif, pH_norm, pH_exp]:
    y = arr.mean(axis=0)
    err = arr.std(axis=0)/np.sqrt(arr.shape[0])
    ax.errorbar(x, y, yerr=err)

ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('total # samples')
ax.set_ylabel('p(choose H)')
ax.set_title('probability of getting H given uniform expiration')
plt.show()
```

```python
alldata = load_from_mysql(dburl="mysql://lab:fishneversink@gureckislab.org/mt_experiments", 
                          tablename='battleship_infounc_2', 
                          codeversion="11.0")

workerids = sort(alldata.keys())
DATA = []
for wid in workerids:
    DATA.append(alldata[wid])
SUBJ = range(len(DATA))
```

```python
for sid in SUBJ:
    s = Subject(sid)
    print '\nSubject %s' % sid
    print 'Planning condition %s' % s.condition_tbt
    print 'Expiration condition %s' % s.condition_exp
    print 'Final bonus %s' % s.bonus
    print 'Expirations: ', [s.games[i].expiration for i in range(N_GAMES)]
    print '# of samples: ', [s.games[i].n_samples for i in range(N_GAMES)]
    print 'Observed expiration: ', [s.games[i].observed_expiration for i in range(N_GAMES)]
```

```python
# create dataframe
rows = []
for s in [Subject(sid) for sid in SUBJ]:
    mn_num_samples = np.mean([s.games[i].n_samples for i in range(N_GAMES) if s.games[i].observed_expiration is None])
    md_num_samples = np.median([s.games[i].n_samples for i in range(N_GAMES) if s.games[i].observed_expiration is None])
    prop_chose_H = np.sum([s.games[i].chose_H for i in range(N_GAMES)])/float(N_GAMES)
    
    nonexp_games = filter(lambda i: s.games[i].observed_expiration==None, range(N_GAMES))
    prop_chose_H_nonexp = np.sum([s.games[i].chose_H for i in nonexp_games])/float(len(nonexp_games))
    
    rows.append([s.sid, 
                 rawdata(s.sid)['workerId'], 
                 int(s.questiondata['age']),
                 int(s.questiondata['difficulty']),
                 int(s.questiondata['engagement']),
                 s.questiondata['gender'],
                 s.condition, 
                 s.condition_tbt, 
                 s.condition_exp, 
                 s.bonus, 
                 mn_num_samples,
                 md_num_samples,
                 prop_chose_H,
                 prop_chose_H_nonexp])
    
df = pd.DataFrame(rows, columns=['sid', 
                                 'workerId', 
                                 'age',
                                 'difficulty',
                                 'engagement',
                                 'gender',
                                 'cond', 
                                 'cond_tbt', 
                                 'cond_exp', 
                                 'bonus', 
                                 'mn_n_samples', 
                                 'md_n_samples',
                                 'prop_chose_H',
                                 'prop_chose_H_nonexp'])
print df.to_string()
print

counts = []
for ci in range(2):
    for cj in range(4):
        counts.append([ci, cj, df[(df['cond_tbt']==ci) & (df['cond_exp']==cj)].shape[0]])
print pd.DataFrame(counts, columns=['cond_tbt', 'cond_exp', 'N'])

bonuses = np.sum([max(0, b) for b in df['bonus']])
fees = len(SUBJ)

print 'Total N: %s' % len(SUBJ)
print 'Fees: %s' % fees
print 'Bonuses: %s' % bonuses
print 'Total cost: %s' % (fees + bonuses)
np.unique([rawdata(sid)['workerId'] for sid in SUBJ]).shape
```

```python
rows = []
for s in [Subject(sid) for sid in SUBJ]:
    for i in range(N_GAMES):
        
        rows.append([s.sid,
                     s.condition_tbt,
                     s.condition_exp,
                     i,
                     s.games[i].n_samples,
                     False if s.games[i].observed_expiration==None else True,
                     s.games[i].chose_H])
        
        dfByGame = pd.DataFrame(rows, columns=['sid',
                                               'cond_tbt',
                                               'cond_exp',
                                               'gameind',
                                               'n_samples',
                                               'observed_exp',
                                               'chose_H'])
print dfByGame.to_string()
print
```
