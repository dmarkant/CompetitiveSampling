```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import requests, json
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style("white") 

from scipy import stats
import statsmodels.api as sm

from mypy.explib.compsamp import rttb_v0 as exp

FIGDEST = '/Users/markant/code/CompetitiveSampling/doc/learningtocompete/figures/'


def get_chooser(sid0, sid1, gid):
    result = []
    for sid in [sid0, sid1]:
        dat = exp.Subject(sid).data
        start = map(lambda d: d[0]=='game' and d[1]==gid, dat).index(True)
        if gid is 7:
            end = -1
        else:
            end = map(lambda d: d[0]=='game' and d[1]==(gid+1), dat).index(True)
        other = len(filter(lambda d: len(d) > 1 and d[1]=='You both stopped, but the other player was randomly selected to choose first...', dat[start:end]))
        chooser = not other
        result.append(chooser)
    return result
```

```python
exp.download_data_from_keyfile(quiet=False, force=True)
```

```python
SUBJ = exp.subjects()
subjects = {sid: exp.Subject(sid) for sid in SUBJ}
SUBJ = filter(lambda sid: np.min(subjects[sid].samplesize()) > 0, SUBJ) # excluding participants with ss=0

cond = [subjects[sid].CONDITION for sid in SUBJ]

subj_by_cond = [[], []]
for i, sid in enumerate(SUBJ):
    subj_by_cond[cond[i]].append(sid)

cond_counts = {}
for c in set(cond):
    cond_counts[c] = cond.count(c)
    
print('N (total):\t\t%i' % len(SUBJ))
for c in set(cond):
    print('N (condition %i):\t%i' % (c, cond_counts[c]))
```

## Competitive conditions

```python
#! currently leaving out any incomplete pairs
sset = SUBJ
uids = [subjects[sid].pars['USER_ID'] for sid in sset]
pairs = []
for sid in sset:
    opp_id = subjects[sid].opponent()
    if opp_id in uids:
        p = [sid, sset[uids.index(opp_id)]]
        p.sort()
        if p not in pairs:
            pairs.append(p)
    else:
        pass
        #print 'opponent not found'
```

```python
arr = []
for pid, pair in enumerate(pairs):
    arr.append(np.transpose((np.ones(exp.N_GAMES) * pid,
                             np.ones(exp.N_GAMES) * pair[0],
                             np.ones(exp.N_GAMES) * pair[1],
                             np.ones(exp.N_GAMES) * subjects[pair[0]].CONDITION,
                             range(exp.N_GAMES),
                             np.min(np.transpose([subjects[sid].samplesize() for sid in pair]), axis=1),
                             np.mean(np.transpose([subjects[sid].samplesize() for sid in pair]), axis=1),
                             subjects[pair[0]].samplesize(),
                             subjects[pair[1]].samplesize(),
                             subjects[pair[0]].stopped(),
                             subjects[pair[1]].stopped(),
                             [exp.Subject(pair[0]).choice(gid) for gid in range(exp.N_GAMES)],
                             [exp.Subject(pair[1]).choice(gid) for gid in range(exp.N_GAMES)],
                             subjects[pair[0]].loss(),
                             subjects[pair[1]].loss(),
                             subjects[pair[0]].loss_random()
                             )))
arr = np.array(arr)
arr = np.array(arr).reshape((arr.shape[0]*arr.shape[1], arr.shape[2]))
pairdf = pd.DataFrame(arr, columns=['pairid', 'sid0', 'sid1', 'cond', 'gameind', 
                                    'minsamplesize', 'avgsamplesize', 'ss0', 'ss1', 
                                    'stop0', 'stop1',
                                    'H_0', 'H_1',
                                    'loss0', 'loss1', 'lossrand'])
pairdf['0stopsfirst'] = pairdf['ss0'] < pairdf['ss1']
pairdf['1stopsfirst'] = pairdf['ss1'] < pairdf['ss0']
pairdf['tie'] = 1*((pairdf.stop0 + pairdf.stop1)==2)

chooser = []
chooserloss = []
chooserH = []
for i, r in pairdf.iterrows():
    if r.cond == 0:
        chooser.append(np.nan)
        chooserloss.append(np.nan)
        chooserH.append(np.nan)
    else:
        if r['stop0']==1 and r['stop1']==0:
            chooser.append(0)
            chooserloss.append(r['loss0'])
            chooserH.append(r['H_0'])
        elif r['stop0']==0 and r['stop1']==1:
            chooser.append(1)
            chooserloss.append(r['loss1'])
            chooserH.append(r['H_1'])
        else:
            # a tie!
            #print map(int, [r['sid0'], r['sid1'], r['gameind']])
            c = get_chooser(int(r['sid0']), int(r['sid1']), int(r['gameind']))
            if c[0]:
                chooser.append(0)
                chooserloss.append(r['loss0'])
                chooserH.append(r['H_0'])
            else:
                chooser.append(1)
                chooserloss.append(r['loss1'])
                chooserH.append(r['H_1'])
pairdf['chooser'] = chooser
pairdf['chooserloss'] = chooserloss
pairdf['chooserH'] = chooserH
```
