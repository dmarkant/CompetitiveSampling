```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def estimated_best_duration(gamma, est_p_success):
    """
    gamma: sensitivity to changes in payoffs
    est_p_success: predicted probability of success (stationary)
    """
    return -gamma / log(est_p_success)
```

### Evaluation

Prior to each round, model identifies how many trials it should play in order to maximize expected gains.

Sensitivity to changes in payoffs ($gamma^+$) will influence how long model "plans" to sample. High sensitivity means longer sampling:

```python
fig = plt.figure()
plt.plot([estimated_best_duration(.5, p) for p in np.arange(0, 1, .1)])
plt.plot([estimated_best_duration(1, p) for p in np.arange(0, 1, .1)])
plt.plot([estimated_best_duration(2, p) for p in np.arange(0, 1, .1)])
plt.show()
```

### Learning

During the BART, people may have incorrect belief that the probability of success is stationary. The BSR allows for either stationary or non-stationary beliefs.

#### Stationary

Prior modeled as beta dstribution $f(q)$, with the mean of the distribution the estimated subjective probability (used above). Beliefs are updated using Bayes' rule after each round. 

#### Non-stationary

In actuality, probability of success is decreasing over trials. Since the stopping point is a randomly selected trial, the probability distribution depends on the total number of trials (but in some cases this is unknown to the participant).

Again use a beta distribution to represent beliefs about the maximum number of samples.
