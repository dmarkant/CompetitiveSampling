```python
# say range of means is drawn uniformly between [20, 40]
mu_rng = [20, 40]

# say range of outcomes is truncated between [0, 100]
x_rng = [0, 100]

x = np.arange(x_rng[0], 1+x_rng[1])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8,2))
y = [weibull_min.pdf(xi, 3, loc=10, scale=40) for xi in x]
ax[0].plot(x, y, 'k')
ax[0].fill_between(x, y, 0, color='.92')

y = [weibull_min.pdf(xi, 3, loc=30, scale=20) for xi in x]
ax[0].plot(x, y, 'k')
ax[0].fill_between(x, y, 0, color='.80')
ax[0].set_ylim(0, .1)
ax[0].set_xlim(0, 100)

y = [weibull_min.pdf(xi, 2, loc=0, scale=40) for xi in x]
ax[1].plot(x, y, 'k')
ax[1].fill_between(x, y, 0, color='.92')

y = [weibull_min.pdf(xi, 4, loc=0, scale=40) for xi in x]
ax[1].plot(x, y, 'k')
ax[1].fill_between(x, y, 0, color='.80')
ax[1].set_ylim(0, .05)
ax[1].set_xlim(0, 100)

plt.show()
```

Show for each option the learning curve.

Then the loss function is defined by pairing of two options.

```python
d1 = gumbel_r(40, 20)
d2 = gumbel_r(40, 30)
d3 = gumbel_r(40, 40)
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(d1.ppf(0.01), d2.ppf(0.99), 100)
ax.plot(x, d1.pdf(x), 'r-', lw=5, alpha=0.6)
ax.plot(x, d2.pdf(x), 'b-', lw=5, alpha=0.6)
ax.plot(x, d3.pdf(x), 'g-', lw=5, alpha=0.6)
```

```python
d1 = gumbel_r(40, 30)
d2 = gumbel_r(50, 50)
d3 = gumbel_r(60, 20)

fig, ax = plt.subplots(1, 1)
x = np.linspace(d1.ppf(0.01), d2.ppf(0.99), 100)
ax.plot(x, d1.pdf(x), 'r-', lw=5, alpha=0.6)
ax.plot(x, d2.pdf(x), 'b-', lw=5, alpha=0.6)
ax.plot(x, d3.pdf(x), 'g-', lw=5, alpha=0.6)
ax.set_xlim(0, 250)
```

```python
def err(distr):
    m = distr.stats(moments='m')
    err = np.array([np.abs([m - np.mean(distr.rvs(size=s)) for s in range(1,100)]) for i in range(100)])
    return err
```

```python
d1err = np.mean(err(d1), axis=0)
d2err = np.mean(err(d2), axis=0)
d3err = np.mean(err(d3), axis=0)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(d1err, 'r-', lw=5, alpha=0.6)
ax.plot(d2err, 'b-', lw=5, alpha=0.6)
ax.plot(d3err, 'g-', lw=5, alpha=0.6)
ax.set_xlim(0, 50)
```

Okay so now consider if you have 2 options that come from different distributions. 

```python
d1 = gumbel_r(40, 10)
d1m, d1v = d1.stats(moments='mv')
d2 = gumbel_r(35, 25)
d2m, d2v = d2.stats(moments='mv')

fig, ax = plt.subplots(1, 2)
x = np.linspace(0, 100)
ax[0].plot(x, d1.pdf(x), 'r-', lw=5, alpha=0.6)
ax[0].plot([d1m, d1m], [0, d1.pdf(d1m)], '-r')
ax[0].plot(x, d2.pdf(x), 'b-', lw=5, alpha=0.6)
ax[0].plot([d2m, d2m], [0, d2.pdf(d2m)], '-b')
ax[0].set_xlim(0, 100)

#ax[1].plot(d1err, 'r-', lw=5, alpha=0.6)
ax[1].plot([np.sqrt(d1v)/np.sqrt(n) for n in range(1,50)], 'r--', lw=2)

#ax[1].plot(d2err, 'b-', lw=5, alpha=0.6)
ax[1].plot([np.sqrt(d2v)/np.sqrt(n) for n in range(1,50)], 'b--', lw=2)
ax[1].set_xlim(0, 50)
```

```python
d1sd = np.array([np.sqrt(d1v)/np.sqrt(n) for n in range(1,50)])
d2sd = np.array([np.sqrt(d2v)/np.sqrt(n) for n in range(1,50)])
```

```python
x = range(1, 50)
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, [d1m for _ in x], 'r-', lw=2, alpha=.7)
ax[0].fill_between(x, d1m + d1sd/2, d1m - d1sd/2, color='red', alpha=.2)
ax[0].plot(x, [d2m for _ in x], 'b-', lw=2, alpha=.7)
ax[0].fill_between(x, d2m + d2sd/2, d2m - d2sd/2, color='blue', alpha=.2)
ax[0].set_xlim(0, 50)
plt.show()
```
