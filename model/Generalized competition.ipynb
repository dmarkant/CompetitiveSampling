```python
V_0 = 0.3
r = .15
def exp_decay(t):
    return V_0 * np.exp(-r * t)

sc = 0.3
b = 15
def sigmoid(t):
    return V_0 - V_0 * (1 / (1 + np.exp(sc * (b - t))))


nsamp = np.arange(0, 30)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(nsamp, [exp_decay(t) for t in nsamp])
ax.plot(nsamp, [sigmoid(t) for t in nsamp])
    
ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('# of samples')
ax.set_ylabel('error')
ax.set_title('Learning curve')
plt.show()
```

## Effect of competition

What might be known about competitors?

1. Number of competitors. Gives the ratio of competitors to options.

2. Shape of the choice function

Given:

- sample size N
- learning curve
- distribution of options
- number of competitors C
- consumption function

How does the consumption function F work?

For each competitor, decide according to F if they stop on trial N. If they stop, sample a preference ordering based on the sample size (for low samples, these will tend to be noisy; as sample size increases, they will be more similar across competitors). 

Simulate stopping round. Given number of stopping competitors S, enumerate possible orderings (including self). 

For each possible ordering, simulate stop decision. Each player attempts to consume option with highest perceived value. If it is taken, they move down until they find an option that is not taken. Record the value of the option that is 


What is the best trial on which to stop?


```python
#est_value_error = sigmoid
est_value_error = exp_decay
```

```python
def sim_obtained_value(M=2, C=2, target_trial=0, option_env='normal', comp_env='uniform', learning='exp'):
    
    if learning is 'exp':
        est_value_error = exp_decay
    elif learning is 'sigmoid':
        est_value_error = sigmoid

    
    options = generate_option_values(M, option_env)
    stoptrials = np.random.randint(1, NMAX, C)
    #print 'sampled stopping trials:', stoptrials
    
    # resolve any earlier stopping decisions to get current option set
    consumed = []
    
    for trial in range(1, target_trial):
        n_stopping = np.sum(stoptrials==trial)
        #print '# people stopped on this trial:', n_stopping
        
        # get preference orders for all stopping competitors
        pref_orders = []
        for i in range(n_stopping):
            est_value = np.array([np.random.normal(loc=opt, scale=est_value_error(trial)) for opt in options])
            pref_orders.append((-est_value).argsort())
    
        #print pref_orders
            
        # get choice order for competitors
        choice_order = range(n_stopping)
        
        # for each chooser, go through ranked preferences and 
        # consume the first one that is available
        for chooser in choice_order:
            
            done = False
            i = -1
            while not done:
                i += 1
                if pref_orders[chooser][i] not in consumed:
                    consumed.append(pref_orders[chooser][i])
                    done = True  
                    
        #print consumed
    
    # now for target trial, do the same while including self
    n_stopping = np.sum(stoptrials==target_trial)
    
    # get preference orders for all stopping competitors, including self
    pref_orders = []
    
    own_est_value = np.array([np.random.normal(loc=opt, scale=est_value_error(target_trial)) for opt in options])
    own_pref_order = (-own_est_value).argsort()
    pref_orders.append(own_pref_order)
    
    for i in range(n_stopping):
        est_value = np.array([np.random.normal(loc=opt, scale=est_value_error(target_trial)) for opt in options])
        pref_orders.append((-est_value).argsort())
    
    
    # sample a choice order for self (0) and competitors
    # (todo) shuffle
    choice_order = range(n_stopping + 1)
    
    obtained_value = None
    for chooser in choice_order:
        
        done = False
        i = -1
        while not done:
            i += 1
            option_index = pref_orders[chooser][i]
            if option_index not in consumed:
                consumed.append(option_index)
                done = True
                
                if chooser == 0:
                    obtained_value = options[option_index]
    
    #print consumed
    return np.max(options) - obtained_value
```
