```python
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, beta
import numpy as np
```

Distribution of options in the environment

```python
x = np.arange(0, 1, .05)

fig, ax = plt.subplots(1, 3, figsize=(10,3))
ax[0].plot(x, np.array([1/len(x) for i in x]))
ax[1].plot(x, np.array([norm.pdf(i, loc=.5, scale=.2) for i in x]))
ax[2].plot(x, np.array([beta.pdf(i, 1, 15) for i in x]))
for i in range(3):
    ax[i].set_xlabel('value')
    ax[i].set_ylabel('p(value)')
    ax[i].set_title('Probability of expiring')

ax[2].set_xlabel('value')
ax[2].set_ylabel('p(value)')
ax[2].set_title('Probability of expiring')
```

## Information uptake function

```python
V_0 = 1
r = .3
def est_value_error(t):
    return V_0 * np.exp(-r * t)

def sigmoid(t):
    return 1 / (1 + np.exp(t))

nsamp = np.arange(0, 10)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(nsamp, [est_value_error(t) for t in nsamp])
ax.plot(nsamp, [sigmoid(t) for t in nsamp])
    

ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('trial')
ax.set_ylabel('p(expire)')
ax.set_title('Probability of expiring')
```
