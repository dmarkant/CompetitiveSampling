```python
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, beta
import numpy as np
```

Distribution of options in the environment

```python
x = np.arange(0, 1, .05)

fig, ax = plt.subplots(1, 3, figsize=(10,3))
ax[0].plot(x, np.array([1/len(x) for i in x]))
ax[1].plot(x, np.array([norm.pdf(i, loc=.5, scale=.2) for i in x]))
ax[2].plot(x, np.array([beta.pdf(i, 1, 15) for i in x]))
for i in range(3):
    ax[i].set_xlabel('value')
    ax[i].set_ylabel('p(value)')
    ax[i].set_title('Distribution of options')
plt.show()
```

## Information uptake function

what is the starting error? use the mean of the distribution as the initial estimate, which means that the initial error will be the standard deviation?

```python
for sc in [0.1, 0.2, 0.3, 0.4]:
    err = np.mean(np.abs(0.5 - np.random.normal(loc=0.5, scale=sc, size=1000000)))
    print sc, err
```

```python
V_0 = 0.5
r = .1
def est_value_error(t):
    return V_0 * np.exp(-r * t)

sc = 0.4
b = 15
def sigmoid(t):
    return V_0 - V_0 * (1 / (1 + np.exp(sc * (b - t))))

nsamp = np.arange(0, 30)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(nsamp, [est_value_error(t) for t in nsamp])
ax.plot(nsamp, [sigmoid(t) for t in nsamp])
    
ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('# of samples')
ax.set_ylabel('error')
ax.set_title('Information uptake')
plt.show()
```
