A choice set $G$ is composed of a high ($H$) option and a low ($L$) option. 

$H: (H_+, H_-, H_p)$

$L: (L_+, L_-, L_p)$

For a given number of samples allocated to a single option ($N_H$ or $N_L$), the probability of observing some frequency of the positive outcomes ($n_{+}$) is given by the binomial distribution. Since there are two options, the probability of the observed frequencies of positive outcomes from each $(n_{H+}, n_{L+})$ is the joint binomial distribution.

$p(n_{H+}) \sim Binom(N_H, H_p)$

$p(n_{L+}) \sim Binom(N_L, L_p)$

And the observed average value for each option is:

$M_H = n_{H+} H_+ + (N_H - n_{H+}) H_-$

$M_L = n_{L+} L_+ + (N_L - n_{L+}) L_-$

We begin by assuming a simple choice rule, that an observer will pick the option with the highest average value based on their samples. The probability of choosing the $H$ option thus depends on the joint probability of observing pairs of values for which the $H$ option is higher. For example, for $N_H=5$ and $N_L=5$, the following shows the joint probability and the difference in the observed mean value given the frequencies of the positive outcome for each option. 

```python
N_H = 5
N_L = 5

x = np.arange(0, N_H + 1, 1)
y = np.arange(0, N_L + 1, 1)
X, Y = np.meshgrid(x, y)

# observed means
Z = np.zeros((N_H + 1, N_L + 1))
for i, j in np.nditer([X, Y]):
    Z[i, j] = value_given_outcomes(g['H'], i, N_H-i) - value_given_outcomes(g['L'], j, N_L-j)
    
# joint probability
Zp = np.zeros((N_H + 1, N_L + 1))
for i, j in np.nditer([X,Y]):
    Zp[i, j] = binom.pmf(i, N_H, g['H'][2]) * binom.pmf(j, N_L, g['L'][2])
    
fig, ax = plt.subplots(1, 2, figsize=(8,8))
ax[0].contourf(X, Y, np.transpose(Zp), 10,
                        cmap='gray',
                        origin='lower')
ax[0].set_xlabel('$n_{H+}$', fontsize=16)
ax[0].set_ylabel('$n_{L+}$', fontsize=16)
ax[0].set_aspect('equal')
ax[0].set_title('Joint probability')

ax[1].contourf(X, Y, np.transpose(Z), 10,
                        cmap='RdBu_r',
                        origin='lower')
ax[1].set_xlabel('$n_{H+}$', fontsize=16)
ax[1].set_ylabel('$n_{L+}$', fontsize=16)
ax[1].set_aspect('equal')
ax[1].set_title('difference in observed value\n(blue=negative, red=positive)')

fig.subplots_adjust(wspace=.2)
plt.show()
```

Thus, the probability of selecting the $H$ option is the summed joint probability over the region of this space where $M_H > M_L$. Note there are two special cases in the choice rule. First, if no samples were allocated to one of the options (and means can't be compared), a simple positive-take, negative-reject choice rule is used. Second, if it happens that the observed mean of the two options is the same, an option is chosen at random ($p(H)=.5$).

It's possible that people plan to sample each option a certain number of times, but we won't commit to that just yet. Instead, we will assume that people plan the total number of samples they will take, and the model will average across all possible allocations to get the overall probability of choosing the $H$ option.

```python
N = 10
prob_choose_H_all_allocations(g, 10)
```

For a single gamble, can now see how the number of samples increases the chance of selecting the H option.

```python
maxsamp = 25
pH = np.array([prob_choose_H_all_allocations(g, n) for n in range(1, maxsamp)])

ev_random = 0.5 * expected_value(g['H']) + 0.5 * expected_value(g['L'])
eg = (pH * expected_value(g['H']) + (1 - pH) * expected_value(g['L'])) - ev_random

maxsamp_per_option = 12
ediff = expected_difference_over_sample_size(g, maxsamp_per_option)

fig, ax = plt.subplots(1, 3, figsize=(12,8))
ax[0].plot(range(1,maxsamp), pH)
ax[0].set_title('Probability of choosing H')
ax[0].set_xlabel('total # samples')
ax[0].set_ylabel('p(get H)')
ax[0].set_aspect(65)

ax[1].plot(range(1,maxsamp), eg)
ax[1].set_title('Expected gain relative to random choice')
ax[1].set_xlabel('total # samples')
ax[1].set_ylabel('expected gain')
ax[1].set_aspect(2)

ax[2].plot(range(1,maxsamp_per_option), ediff)
ax[2].set_title('Expected difference between mean values')
ax[2].set_xlabel('# samples per option')
ax[2].set_ylabel('expected difference')
ax[2].set_aspect(.01)

fig.subplots_adjust(wspace=.5)
plt.show()
```
