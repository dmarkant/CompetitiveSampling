## Information uptake function

what is the starting error? use the mean of the distribution as the initial estimate, which means that the initial error will be the standard deviation?

```python
for sc in [0.1, 0.2, 0.3, 0.4]:
    err = np.mean(np.abs(0.5 - np.random.normal(loc=0.5, scale=sc, size=1000000)))
    print sc, err
```

```python
V_0 = 0.3
r = .1
def exp_decay(t):
    return V_0 * np.exp(-r * t)

sc = 0.4
b = 15
def sigmoid(t):
    return V_0 - V_0 * (1 / (1 + np.exp(sc * (b - t))))

nsamp = np.arange(0, 30)
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(nsamp, [exp_decay(t) for t in nsamp])
ax.plot(nsamp, [sigmoid(t) for t in nsamp])
    
ax.grid(b=True, which='major', color='gray', linestyle='-')
ax.set_xlabel('# of samples')
ax.set_ylabel('error')
ax.set_title('Information uptake')
plt.show()
```

Given $N$ options that were generated from distribution $D$

```python
M = 10
niter = 5000
```

```python
def generate_option_values(M, env):
    #options = []
    if env is 'normal':
        result = np.random.beta(4, 4, size=M)
    
    elif env is 'uniform':
        result = np.random.beta(1, 1, size=M)
    
    elif env is 'skewed':
        result = np.random.beta(2, 15, size=M)

    return result
        
    
    #while len(options) < M:
    #    v = dist(loc=0.5, scale=0.2)
    #    if v > 0. and v < 1.:
    #        options.append(v)
            
    #return np.array(options)
        
```

```python
# the true value of the options
options = generate_option_values(M, 'normal')
```

```python
est_value_error = sigmoid
est_value = np.array([[np.random.normal(loc=opt, scale=est_value_error(t)) for t in nsamp] for opt in options])
```

```python
fig, ax = plt.subplots()
for i, opt in enumerate(options):
    ax.plot(nsamp, est_value[i])
```

```python
best = np.max(options)
reward = np.array(options)[np.argmax(np.array(est_value), axis=0)]
loss = best - reward
```

```python
fig, ax = plt.subplots()
ax.plot(nsamp, loss)
```

```python
# fixed costs
cobs = .001
cost = cobs * nsamp
```

```python
est_value_error = exp_decay
loss_exp = np.zeros((niter, len(nsamp)))
for i in range(niter):
    options = generate_option_values(M, 'normal')
    est_value = np.array([[np.random.normal(loc=opt, scale=est_value_error(t)) for t in nsamp] for opt in options])
    best = np.max(options)
    reward = np.array(options)[np.argmax(np.array(est_value), axis=0)]
    loss_exp[i] = best - reward
    
est_value_error = sigmoid
loss_sig = np.zeros((niter, len(nsamp)))
for i in range(niter):
    options = generate_option_values(M, 'normal')
    est_value = np.array([[np.random.normal(loc=opt, scale=est_value_error(t)) for t in nsamp] for opt in options])
    best = np.max(options)
    reward = np.array(options)[np.argmax(np.array(est_value), axis=0)]
    loss_sig[i] = best - reward
```

```python
fig, ax = plt.subplots()
plt.plot(nsamp, np.mean(loss_exp, axis=0))
plt.plot(nsamp, np.mean(loss_exp + cost, axis=0))
plt.plot(nsamp, np.mean(loss_sig, axis=0))
plt.plot(nsamp, np.mean(loss_sig + cost, axis=0))
plt.show()
```
