```python
def estimate(opt, N):
    est_values, probs = option_sample_means_and_probs(opt, N)
    est = np.sum([probs[i] * est_values[i] for i in range(len(est_values))])
    return est
```

```python
def abs_est_error(opt, N):
    ev = expected_value(opt)
    est_values, probs = option_sample_means_and_probs(opt, N)
    #print '\ntrue value:', ev
    #print 'estimate:', est_values
    #print 'probs:\t', probs
    #print 'error:\t', [np.abs(est_values[i] - ev) for i in range(len(est_values))]
    
    w_abs_error = [probs[i] * np.abs(est_values[i] - ev) for i in range(len(est_values))]
    return np.sum(w_abs_error)
```

```python
def error(opt, max_N):
    return np.array([abs_est_error(opt, t) for t in range(1, max_N)])
```

```python
def average_error_for_option_env(n, t, pos_range, neg_range, p_range):
    err = np.zeros((n, t), float)
    for i in range(n):
        opt = generate_option(pos_range=pos_range, neg_range=neg_range, p_range=p_range)
        err[i,:] = error(opt, t+1)
    return err
```

```python
envs = {'env_1': {'pos_range': [20,   40],
                  'neg_range': [-40, -20],
                  'p_range':   [.5, .6]},
        'env_2': {'pos_range': [20,   40],
                  'neg_range': [-250,-20],
                  'p_range':   [.85, .95]}}

NMAX = 50 # maximum trial
n = 100 # iterations

avg_err = {}

for name in envs.keys():
    env = envs[name]
    avg_err[name] = average_error_for_option_env(n, NMAX, 
                                                 env['pos_range'],
                                                 env['neg_range'],
                                                 env['p_range'])
```

```python
fig, ax = plt.subplots(figsize=(15,4))
ax.plot(avg_err['env_1'].mean(axis=0))
ax.plot(avg_err['env_2'].mean(axis=0))
plt.show()
```

```python
def estimated_value(option, n_samples):
    pos, neg, p = option
    if n_samples == 0:
        return 0.
    else:
        n_pos = binomial(n_samples, p)
        v = (n_pos * pos + (n_samples - n_pos) * neg) / float(n_samples)
        return v
```

```python
def sim_obtained_value(M=2, C=2, target_trial=0, option_env='env_1', comp_env='uniform'):
    """Agent-based simulation, returning the obtained loss relative to best option
    
    Keyword args:
    M -- number of options
    C -- number of competitors (in addition to self)
    target_trial -- trial on which loss is evaluated
    """
    assert option_env in avg_err.keys()
    env = envs[option_env]
    
    def est_value_error(t):
        return avg_err[option_env][t]
    
    options = []
    for _ in range(M):
        options.append(generate_option(pos_range=env['pos_range'],
                                       neg_range=env['neg_range'],
                                       p_range=env['p_range']))
    
    expected_values = [expected_value(opt) for opt in options]
    
    #print 'true expected values:', expected_values
    
    stoptrials = np.random.randint(1, NMAX, C)
    
    # resolve any earlier stopping decisions to get current option set
    consumed = []
    
    for trial in range(1, target_trial):
        
        # number of people stopping on this trial
        n_stopping = np.sum(stoptrials==trial)
        
        # get preference orders for all stopping competitors
        pref_orders = []
        for i in range(n_stopping):
            
            # assign sampling trials to options
            sampled_options = np.random.choice(range(len(options)), trial)
 
            # count samples per option
            n_sample_by_option = [np.sum(sampled_options==i) for i in range(len(options))]
            
            # get estimated value of each option based on # of samples
            est_value = np.array([estimated_value(options[i], n_sample_by_option[i]) for i in range(len(options))])
            pref_orders.append((-est_value).argsort())
            
        # get choice order for competitors
        choice_order = range(n_stopping)
        shuffle(choice_order)
        
        # for each chooser, go through ranked preferences and 
        # consume the first one that is available
        for chooser in choice_order:
            
            done = False
            i = -1
            while not done:
                i += 1
                if pref_orders[chooser][i] not in consumed:
                    consumed.append(pref_orders[chooser][i])
                    done = True  
    
    # now for target trial, do the same while including self
    n_stopping = np.sum(stoptrials==target_trial)
    
    # get preference orders for all stopping competitors, including self
    pref_orders = []
    
    # assign sampling trials to options
    own_sampled_options = np.random.choice(range(len(options)), target_trial)
            
    # count samples per option
    own_n_sample_by_option = [np.sum(own_sampled_options==i) for i in range(len(options))]
            
    # get estimated value of each option based on # of samples
    own_est_value = np.array([estimated_value(options[i], own_n_sample_by_option[i]) for i in range(len(options))])
    own_pref_order = (-own_est_value).argsort()
    pref_orders.append(own_pref_order)
        
    for i in range(n_stopping):
        
        # assign sampling trials to options
        sampled_options = np.random.choice(range(len(options)), target_trial)

        # count samples per option
        n_sample_by_option = [np.sum(sampled_options==i) for i in range(len(options))]
        
        # get estimated value of each option based on # of samples
        est_value = np.array([estimated_value(options[i], n_sample_by_option[i]) for i in range(len(options))])
        pref_orders.append((-est_value).argsort())
    
    
    # sample a choice order for self (0) and competitors
    choice_order = range(n_stopping + 1)
    shuffle(choice_order)
    
    obtained_value = None
    for chooser in choice_order:
        
        done = False
        i = -1
        while not done:
            i += 1
            option_index = pref_orders[chooser][i]
            if option_index not in consumed:
                consumed.append(option_index)
                done = True
                
                if chooser == 0:
                    obtained_value = expected_value(options[option_index])
    
    
    #print 'option values:', expected_values
    #print 'obtained:', obtained_value
    best = np.max([expected_value(opt) for opt in options])
    #print 'best:', best
    
    return best - obtained_value
```

```python
NMAX = 30
dN = 1
t_dim = range(0, NMAX, dN)
```

```python
niter = 5000
sim_result = {}
```

```python
def iter_obtained_value(ntrials, M=2, C=2, option_env='env_1', comp_env='uniform'):
    v = np.zeros((niter, ntrials), float)
    for i in range(niter):
        v[i,:] = np.array([sim_obtained_value(M=M, C=C, target_trial=t, option_env=option_env) for t in range(ntrials)])
    return v.mean(axis=0)
```
